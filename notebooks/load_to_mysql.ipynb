```python
import pandas as pd
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Aryan@2010",
    database="movie_recommender"
)

cursor = conn.cursor()
print("✅ Connected to MySQL!")
```

```python
# Bronze = raw data, exactly as downloaded

cursor.execute("""l
    CREATE TABLE IF NOT EXISTS bronze_movies (
        movieId INT,
        title VARCHAR(255),
        genres VARCHAR(500)
    )
""")

cursor.execute("""
    CREATE TABLE IF NOT EXISTS bronze_ratings (
        userId INT,
        movieId INT,
        rating FLOAT,
        timestamp BIGINT
    )
""")

cursor.execute("""
    CREATE TABLE IF NOT EXISTS bronze_tags (
        userId INT,
        movieId INT,
        tag VARCHAR(255),
        timestamp BIGINT
    )
""")

conn.commit()
print("✅ Bronze tables created!")
```

```python
# Load CSVs
movies  = pd.read_csv('../data/raw/movies.csv')
ratings = pd.read_csv('../data/raw/ratings.csv')
tags    = pd.read_csv('../data/raw/tags.csv')

# Insert movies
for _, row in movies.iterrows():
    cursor.execute(
        "INSERT INTO bronze_movies VALUES (%s, %s, %s)",
        tuple(row)
    )

# Insert ratings
for _, row in ratings.iterrows():
    cursor.execute(
        "INSERT INTO bronze_ratings VALUES (%s, %s, %s, %s)",
        tuple(row)
    )

# Insert tags
for _, row in tags.iterrows():
    cursor.execute(
        "INSERT INTO bronze_tags VALUES (%s, %s, %s, %s)",
        tuple(row)
    )

conn.commit()
print("✅ Raw data loaded into Bronze layer!")
```

```python
cursor.execute("SELECT COUNT(*) FROM bronze_movies")
print("Movies in Bronze:", cursor.fetchone()[0])

cursor.execute("SELECT COUNT(*) FROM bronze_ratings")
print("Ratings in Bronze:", cursor.fetchone()[0])

cursor.execute("SELECT COUNT(*) FROM bronze_tags")
print("Tags in Bronze:", cursor.fetchone()[0])
```
